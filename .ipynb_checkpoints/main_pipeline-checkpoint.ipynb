### Importing Required Libraries

```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

### GETTING CALIBRATION PARAMETERS

```python
#Chessboard Size
nx, ny = 9, 6

#Creating Placeholders
objpoints = [] #3D points in "real world" space
imgpoints = [] #2D points in image plane

# Prepare object points
objp = np.zeros((nx*ny,3), np.float32)
objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2) #x,y coords

# Reading images
checkerboards = glob.glob('./camera_cal/calibration*.jpg')
img1 = mpimg.imread(sorted(checkerboards)[0])
imshape = img1.shape
detected_corners = []

# Iterating images
for fname in checkerboards:
    img = mpimg.imread(fname) #read in each image
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) #convert image to grayscale
    ret, corners = cv2.findChessboardCorners(gray, (nx,ny), None) #find the chessboard corners
        
    # If corners found: Add object points, image points
    if ret==True:
        imgpoints.append(corners)
        objpoints.append(objp)
        # Drawing corners (optional)
        drawCb = cv2.drawChessboardCorners(img, (nx,ny), corners, ret)
        detected_corners.append(drawCb)

# Calibrating camera and finding the camera matrix and the distortion coefficients
ret, cameraMatrix, distortionCoeffs, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, imshape[:2],None,None)
```

### PIPELINE

#### Pre-Processing the Image

```python
## PREPARING BINARY WARPED IMAGE
################################
def get_binary_warped(img):   
    
    ## PERSPECTIVE TRANSFORM
    ########################
    def perspective_transform(thresholded):
        # Image Size
        h,w = thresholded.shape[0], thresholded.shape[1]
        # The coordinates of quadrangle vertices in the source image
        s1,s2,s3,s4 = [w // 2 - 75, h*0.625], [w // 2 + 75,h*0.625], [-100,h], [w+100,h]
        img_in = np.float32([s1,s2,s3,s4])
        # The coordinates of quadrangle vertices in the target image
        t1,t2,t3,t4 = [100,0], [w-100,0], [100,h], [w-100,h]
        img_tar = np.float32([t1,t2,t3,t4])
        # The perspective transform matrix
        Mpt = cv2.getPerspectiveTransform(img_in, img_tar)
        # The inverse perspective transform matrix
        Mpti = cv2.getPerspectiveTransform(img_tar, img_in)
        # Warping the image
        img_warped = cv2.warpPerspective(thresholded, Mpt, (w,h), flags=cv2.INTER_LINEAR)
        # Unwarping
        unwarp = cv2.getPerspectiveTransform(img_in,img_tar)
        return img_warped,unwarp,Mpt,Mpti

    ## COMBINED GRADIENT & COLOR TRANSFORM
    ######################################
    def abs_sobel_thresh(gray, orient='x', thresh_min=0, thresh_max=255):
        if orient == 'x':
            sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
        else:
            sobel = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
        abs_sobel = np.absolute(sobel)
        max_value = np.max(abs_sobel)
        binary_output = np.uint8(255*abs_sobel/max_value)
        threshold_mask = np.zeros_like(binary_output)
        threshold_mask[(binary_output >= thresh_min) & (binary_output <= thresh_max)] = 1
        return threshold_mask

    def dir_threshold(gray, sobel_kernel=3, thresh=(0, np.pi/2)):
        # Horizontal and Vertical Gradients
        abs_sobel_x = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
        abs_sobel_y = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
        # Calculating the direction of the gradient
        direction = np.arctan2(abs_sobel_y,abs_sobel_x)
        direction = np.absolute(direction)
        # Creating binary image in the threshold boundaries
        mask = np.zeros_like(direction)
        mask[(direction >= thresh[0]) & (direction <= thresh[1])] = 1
        return mask

    def get_thresholded(gray):
        # Image Shape
        h,w = gray.shape
        # Gradient threshold on horizontal gradient
        thresh_min, thresh_max = 55, 180
        sobel_x = abs_sobel_thresh(gray, 'x', thresh_min, thresh_max)
        # Gradient direction threshold to create a binary mask
        thresh_dir=(np.pi/6, np.pi/2)
        dir_binary = dir_threshold(gray, thresh=thresh_dir)
        # Combining the gradients
        combined_condition = ((sobel_x == 1) & (dir_binary == 1))
        # R & G thresholds to detect yellow lanes
        color_threshold = 125
        R,G = img[:,:,0], img[:,:,1]
        color_combined = np.zeros_like(R)
        rg_condition = (R > color_threshold) & (G > color_threshold)
        # HLS thresholds
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        L,S = hls[:,:,1], hls[:,:,2]
        # L channel is performed well for the edges generated due to shadows
        l_thresh = (50, 255)
        l_condition = (L > l_thresh[0]) & (L <= l_thresh[1])
        # S channel to detect white and yellow lanes
        s_thresh = (90, 255)
        s_condition = (S > s_thresh[0]) & (S <= s_thresh[1])
        # Combined color mask to detect whether color of a pixel near to yellow or white
        color_combined[(rg_condition & l_condition) & (s_condition | combined_condition)] = 1
        # Determining the region of interest
        roi = np.zeros_like(color_combined)
        region_of_interest_vertices = np.array([[0,h-1], [w/2, int(0.5*h)], [w-1, h-1]], dtype=np.int32)
        cv2.fillPoly(roi, [region_of_interest_vertices], 1)
        combined_thresholded = cv2.bitwise_and(color_combined, roi)
        return combined_thresholded

    ## UNDISTORTING THE IMAGES
    ##########################
    def get_undistorted(img):
        undistorted = cv2.undistort(img, cameraMatrix, distortionCoeffs, None, cameraMatrix)
        return undistorted
    
    undistorted_img = get_undistorted(img)
    gray = cv2.cvtColor(undistorted_img, cv2.COLOR_RGB2GRAY)
    thresholded = get_thresholded(gray)
    img_warped,unwarp,Mpt,Mpti = perspective_transform(thresholded)
    
    return img_warped,unwarp,Mpt,Mpti
```

```python
# GETTING CALIBRATION PARAMETERS & IMAGE SIZE
img = mpimg.imread('test_images/test2.jpg')
image_shape = img.shape
img_size = (image_shape[1], image_shape[0])
warped,unwarp,Mpt,Mpti = get_binary_warped(img)
```

#### Lane & FS Prediction

```python
# WINDOWING PARAMETERS
num_windows = 10
num_rows = warped.shape[0]
window_height = np.int(num_rows/num_windows)
window_half_width = 50
min_pixels = 100

# VARIABLES FOR BUFFER
polyfit_l, polyfit_r = None, None
buffer_lines_l = []
buffer_lines_r = []
current_avg_dif_btw_lines = 0

## MEASURING CURVATURE
######################
def measure_curvature_real(x_values):
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    # If no pixels were found return None
    y_points = np.linspace(0, num_rows-1, num_rows)
    y_eval = np.max(y_points)
    # Fit new polynomials to x,y in world space
    fit_cr = np.polyfit(y_points*ym_per_pix, x_values*xm_per_pix, 2)
    curverad = ((1 + (2*fit_cr[0]*y_eval*ym_per_pix + fit_cr[1])**2)**1.5) / np.absolute(2*fit_cr[0])
    return curverad

## LANE PIXEL DETECTION: PEAKS AT HISTOGRAM
###########################################
def get_lane_peaks(binary_warped, plotting):
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    if plotting == True:
        plt.plot(histogram)
        plt.show() 
    return leftx_base, rightx_base

## INITIALIZE LINES
###################
def get_line_predictions(non_zeros_x, non_zeros_y, coords_l, coords_r, num_rows):
    # Initial Coordinates
    left_x, left_y = non_zeros_x[coords_l], non_zeros_y[coords_l]
    right_x, right_y = non_zeros_x[coords_r], non_zeros_y[coords_r]
    
    # Return None if no pixel initialized
    if(left_y.size == 0 or left_x.size == 0):
        return None, None
    # Fit the polynomial
    polyfit_l = np.polyfit(left_y, left_x, 2)
    # Return None if no pixel initialized
    if(right_y.size == 0 or right_x.size == 0):
        return None, None
    # Fit the polynomial
    polyfit_r = np.polyfit(right_y, right_x, 2)
    
    # Return None if no pixel initialized
    y_points = np.linspace(0, num_rows-1, num_rows)
    # Generate the lane lines from the polynomial fit
    predictions_lx = polyfit_l[0]*y_points**2 + polyfit_l[1]*y_points + polyfit_l[2]
    predictions_rx = polyfit_r[0]*y_points**2 + polyfit_r[1]*y_points + polyfit_r[2]
    return predictions_lx, predictions_rx

## INITIALIZE PREDICTIONS
#########################
def get_initial_predictions(warped):
    #Creation of base poisitions based on histogram peaks
    non_zeros = warped.nonzero()
    non_zeros_x, non_zeros_y = non_zeros[1], non_zeros[0]
    num_rows = warped.shape[0]
    histogram = np.sum(warped[warped.shape[0]//2:,:], axis=0)
    half_width = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:half_width])
    rightx_base = np.argmax(histogram[half_width:]) + half_width
    # Windowing Parameters
    num_windows = 10
    window_height = np.int(num_rows/num_windows)
    window_half_width = 50
    min_pixels = 100
    coords_l = []
    coords_r = []

    for window in range(num_windows):
        y_max = num_rows - window*window_height
        y_min = num_rows - (window+1)* window_height
        left_x_min = leftx_base - window_half_width
        left_x_max = leftx_base + window_half_width
        passed_win_coords_l = ((non_zeros_x >= left_x_min) & (non_zeros_x <= left_x_max) & (non_zeros_y >= y_min) & (non_zeros_y <= y_max)).nonzero()[0]
        coords_l.append(passed_win_coords_l)
        if len(passed_win_coords_l) > min_pixels:
            leftx_base = np.int(np.mean(non_zeros_x[passed_win_coords_l]))
        right_x_min = rightx_base - window_half_width
        right_x_max = rightx_base + window_half_width
        passed_win_coords_r = ((non_zeros_x >= right_x_min) & (non_zeros_x <= right_x_max) & (non_zeros_y >= y_min) & (non_zeros_y <= y_max)).nonzero()[0]
        coords_r.append(passed_win_coords_r)
        if len(passed_win_coords_r) > min_pixels:
            rightx_base = np.int(np.mean(non_zeros_x[passed_win_coords_r]))
    coords_l = np.concatenate(coords_l)
    coords_r = np.concatenate(coords_r)
    predictions_lx, predictions_rx = get_line_predictions(non_zeros_x, non_zeros_y, coords_l, coords_r, num_rows)
    return predictions_lx, predictions_rx

def get_averaged_line(previous_lines, new_line):
    # Number of frames to average over
    num_frames = 10
    if new_line is None:
        # Return None if no line detected
        if len(previous_lines) == 0:
            return previous_lines, None
        else:
            # Else return the last line
            return previous_lines, previous_lines[-1]
    else:
        if len(previous_lines) < num_frames:
            # If current frames > num_frames do:
            previous_lines.append(new_line)
            return previous_lines, new_line
        else:
            # Take the average for the last 'num_frames'
            previous_lines[0:num_frames-1] = previous_lines[1:]
            previous_lines[num_frames-1] = new_line
            new_line = np.zeros_like(new_line)
            for i in range(num_frames):
                new_line += previous_lines[i]
            new_line /= num_frames
            return previous_lines, new_line
        
def get_avg_distance(left_line, right_line, current_avg):
    # Calculate average distance between lanes
    avg_distance = np.mean(right_line - left_line)
    # Update Average
    if current_avg == 0:
        current_avg = avg_distance
    else:
        current_avg = 0.9*current_avg + 0.1*avg_distance
    return current_avg

def get_processed_frame(img):
    # Import polynomial coefficients for the lines in the last frame
    global polyfit_r
    global polyfit_l
    # Import the line coordinates in the last frames
    global buffer_lines_r
    global buffer_lines_l
    # Import the current average difference between lines
    global current_avg_dif_btw_lines
    # Image Parameters
    img_shape = img.shape
    img_size = (image_shape[1], image_shape[0])
    # Creation of Warped (binary warped) Image
    warped,unwarp,Mpt,Mpti = get_binary_warped(img)
    out_img = np.dstack((warped, warped, warped))*255
    # Initializing non_zeros
    non_zeros = warped.nonzero()
    non_zeros_y = non_zeros[0]
    non_zeros_x = non_zeros[1]
    num_rows = warped.shape[0]
    y_points = np.linspace(0, num_rows-1, num_rows)

    if (polyfit_l is None) or (polyfit_r is None):
        # If no polynomial coefficients found,then initialize polynomial coefficients
        initial_poly_coeff = True
        predictions_lx, predictions_rx = get_initial_predictions(warped)
    else:
        # If found, return predictions for the each side with a given margin
        initial_poly_coeff = False
        margin = 100
        # Predictions for left
        predictions_lx = polyfit_l[0]*non_zeros_y**2 + polyfit_l[1]*non_zeros_y + polyfit_l[2]
        coords_l = ((non_zeros_x >= predictions_lx - margin) & (non_zeros_x <= predictions_lx + margin)).nonzero()[0]
        # Predictions for right
        predictions_rx = polyfit_r[0]*non_zeros_y**2 + polyfit_r[1]*non_zeros_y + polyfit_r[2]
        coords_r = ((non_zeros_x >= predictions_rx - margin) & (non_zeros_x <= predictions_rx + margin)).nonzero()[0]
        # Get predictions
        predictions_lx, predictions_rx = get_line_predictions(non_zeros_x, non_zeros_y, coords_l, coords_r, num_rows)
    # Check whether predictions are failed
    if (predictions_lx is None or predictions_rx is None):
        if not initial_poly_coeff:
            predictions_lx, predictions_rx = get_initial_predictions(warped)        
    failed_lines = False 
    if (predictions_lx is None or predictions_rx is None):
        failed_lines = True
    else:
        mean_difference = np.mean(predictions_rx - predictions_lx)
        if current_avg_dif_btw_lines == 0:
            current_avg_dif_btw_lines = mean_difference
        if (mean_difference < 0.7*current_avg_dif_btw_lines or mean_difference > 1.3*current_avg_dif_btw_lines):
            failed_lines = True
            if not initial_poly_coeff:
                predictions_lx, predictions_rx = get_initial_predictions(warped)
                if (predictions_lx is None or predictions_rx is None):
                    failed_lines = True
                else:
                    mean_difference = np.mean(predictions_rx - predictions_lx)
                    if (mean_difference < 0.7*current_avg_dif_btw_lines or mean_difference > 1.3*current_avg_dif_btw_lines):
                        failed_lines = True
                    else:
                        failed_lines = False
        else:
            failed_lines = False
            
    if failed_lines:
        polyfit_l, polyfit_r = None, None
        if len(buffer_lines_l) == 0 and len(buffer_lines_r) == 0:
            return img
        else:
            predictions_lx = buffer_lines_l[-1]
            predictions_rx = buffer_lines_r[-1]
    else:
        buffer_lines_l, predictions_lx = get_averaged_line(buffer_lines_l, predictions_lx)
        buffer_lines_r, predictions_rx = get_averaged_line(buffer_lines_r, predictions_rx)
        mean_difference = np.mean(predictions_rx - predictions_lx)
        current_avg_dif_btw_lines = 0.9*current_avg_dif_btw_lines + 0.1*mean_difference
    #Return line windows for each side
    line_win_l = np.array(np.transpose(np.vstack([predictions_lx, y_points])))
    line_win_r = np.array(np.flipud(np.transpose(np.vstack([predictions_rx, y_points]))))
    # Calculating Radius of Curvature
    curve_rad_l= measure_curvature_real(predictions_lx)
    curve_rad_r= measure_curvature_real(predictions_rx)
    avg_curve_rad = (curve_rad_l + curve_rad_r)/2
    str_curvature = "The radius of curvature: %.2f m" % avg_curve_rad
    # Calculating the Offset from the Center
    lane_center = (predictions_rx[num_rows-1] + predictions_lx[num_rows-1])/2
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    center_offset_pxls = abs(img_size[0]/2 - lane_center)
    center_offset_mtrs = xm_per_pix*center_offset_pxls
    str_offset = "Offset from the center: %.2f m" % center_offset_mtrs
    # Filling the FREE-SPACE between lanes
    poly_points = np.vstack([line_win_l, line_win_r])
    fs_color = [0,255,0] #green
    cv2.fillPoly(out_img, np.int_([poly_points]), fs_color)
    unwarped = cv2.warpPerspective(out_img, Mpti, img_size , flags=cv2.INTER_LINEAR)
    final_frame = cv2.addWeighted(img, 1, unwarped, 0.3, 0)
    # Writing Curvature and Offset onto the Final Image
    txt_offset = (80,80)
    cv2.putText(final_frame, str_curvature, txt_offset, cv2.FONT_HERSHEY_DUPLEX, 1, (255,255,255), thickness=2)
    cv2.putText(final_frame, str_offset, (txt_offset[0],txt_offset[1]+40), cv2.FONT_HERSHEY_DUPLEX, 1, (255,255,255), thickness=2)
    return final_frame
```

```python
# Apply pipeline
final_frame = get_processed_frame(img)

# Plot the 2 images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(img)
ax1.set_title('Original Image', fontsize=20)
ax2.imshow(final_frame, cmap='gray')
ax2.set_title('Final Image', fontsize=20)
plt.subplots_adjust(left=0.2, right=1, top=1, bottom=0)
```
